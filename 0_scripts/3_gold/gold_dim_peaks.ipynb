> ## 1. View Data First

```python
from pyspark.sql import functions as F
```

```python
df = spark.table("himalaya.silver.expeditions_peaks")
```

```python
display(df.limit(5))
```

> ##2. Drop Ingestion

  _not needed in gold_

```python
df = df.drop("ingested_at")
```

>##3. User Friendly Renaming

```python
df = df.withColumnsRenamed({
    "peak_name": "mountain",
    "first_ascent_summiters": "first_ascent_members"
})
```

>##4. Write to Gold

```python
df.write.format("delta").mode("overwrite").saveAsTable("himalaya.gold.dim_peaks")
print("✅ Written to himalaya.gold.dim_peaks")
```
