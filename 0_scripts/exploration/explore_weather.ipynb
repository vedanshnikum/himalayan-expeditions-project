```python
import pyspark.sql.functions as F
```

```python
df = spark.table("himalaya.bronze.weather")
```

```python
# Overview
print(f"Rows:    {df.count()}")
print(f"Columns: {len(df.columns)}")
df.printSchema()
```

```python
# Nulls
print("\nNull counts:")
for c in df.columns:
    null_count = df.filter(F.col(c).isNull()).count()
    status = "✅" if null_count == 0 else f"❌ {null_count}"
    print(f"{status} {c}")
```

```python
# Distinct peaks
print("\nPeaks loaded:")
df.select("peakid").distinct().orderBy("peakid").show()
```

```python
# Distinct weather codes
print("weather codes:")
df.select("weather_code").distinct().show()
```

```python
# Date range
min_date, max_date = df.select(F.min("date"), F.max("date")).first()
print(f"Min date: {min_date}")
print(f"Max date: {max_date}")
```
