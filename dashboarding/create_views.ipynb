# This is for creating views for the dashboard

##Master View - Expeditions
Contains:
* Expeditions
* Peaks
* Weather

```sql
CREATE OR REPLACE VIEW himalaya.gold.vw_master_expeditions AS
SELECT
    f.expid AS `Expedition ID`,
    f.peakid AS `Peak ID`,
    f.weather_id AS `Weather ID`,
    f.year AS `Year`,
    f.season AS `Season`,
    f.host AS `Host Country`,
    f.routes AS `Routes`,
    f.successes AS `Success`,
    f.termreason AS `Termination Reason`,
    f.termnote AS `Termination Note`,
    f.highpoint AS `Highest Point (m)`,
    f.deaths AS `Deaths`,
    f.total_members AS `Total Members`,
    f.summit_members AS `Summit Members`,
    f.total_hired AS `Total Hired`,
    f.summit_hired AS `Summit Hired`,
    f.base_camp_date AS `Base Camp Date`,
    f.summit_date AS `Summit Date`,
    f.termination_date AS `Termination Date`,
    f.camps AS `Camps`,
    f.oxygen_used AS `Oxygen Used`,
    f.accidents AS `Accidents`,
    f.achievement AS `Achievement`,
    f.agency AS `Agency`,
    p.mountain AS `Mountain`,
    p.height_m AS `Height (m)`,
    p.location AS `Location`,
    p.region AS `Region`,
    p.himalayan_range AS `Himalayan Range`,
    p.open AS `Open`,
    p.status AS `Status`,
    p.first_ascent_year AS `First Ascent Year`,
    p.first_ascent_season AS `First Ascent Season`,
    p.first_ascent_country AS `First Ascent Country`,
    p.first_ascent_members AS `First Ascent Members`,
    w.date AS `Weather Date`,
    w.temperature AS `Temperature (°C)`,
    w.rain AS `Rain (mm)`,
    w.snowfall AS `Snowfall (cm)`,
    w.snow_depth AS `Snow Depth (m)`,
    w.wind_speed AS `Wind Speed (km/h)`,
    w.weather_condition AS `Weather Condition`
FROM himalaya.gold.fact_expeditions f
LEFT JOIN himalaya.gold.dim_peaks p ON f.peakid = p.peakid
LEFT JOIN himalaya.gold.dim_weather w ON f.weather_id = w.weather_id;
```

##Master View - Members
Contains:
* Members
* Deaths
* Weather
* Peaks

```sql
CREATE OR REPLACE VIEW himalaya.gold.vw_master_members AS
SELECT
    m.expid AS `Expedition ID`,
    m.peakid AS `Peak ID`,
    m.year AS `Year`,
    m.season AS `Season`,
    m.name AS `Name`,
    m.sex AS `Sex`,
    m.yob AS `Year of Birth`,
    m.nationality AS `Nationality`,
    m.leader AS `Leader`,
    m.hired AS `Hired`,
    m.sherpa AS `Sherpa`,
    m.success AS `Success`,
    m.highest_point AS `Highest Point (m)`,
    m.summit_date AS `Summit Date`,
    m.summit_term AS `Summit Term`,
    m.oxygen_used AS `Oxygen Used`,
    m.death AS `Death`,
    d.cause_of_death_category AS `Cause of Death`,
    d.death_date AS `Death Date`,
    p.mountain AS `Mountain`,
    p.height_m AS `Height (m)`,
    p.region AS `Region`,
    p.himalayan_range AS `Himalayan Range`,
    w.temperature AS `Temperature (°C)`,
    w.rain AS `Rain (mm)`,
    w.snowfall AS `Snowfall (cm)`,
    w.snow_depth AS `Snow Depth (m)`,
    w.wind_speed AS `Wind Speed (km/h)`,
    w.weather_condition AS `Weather Condition`
FROM himalaya.gold.dim_members m
LEFT JOIN himalaya.gold.dim_deaths d ON m.expid = d.expid AND m.name = d.name
LEFT JOIN himalaya.gold.dim_peaks p ON m.peakid = p.peakid
LEFT JOIN himalaya.gold.dim_weather w 
    ON m.peakid = w.peakid 
    AND COALESCE(d.death_date, m.summit_date) = w.date;
```

##Overview

```sql
CREATE OR REPLACE VIEW himalaya.gold.vw_overview AS
SELECT
    `Expedition ID`,
    `Peak ID`,
    `Mountain`,
    `Height (m)`,
    `Region`,
    `Himalayan Range`,
    `Year`,
    `Season`,
    `Success`,
    `Termination Reason`,
    `Deaths`,
    `Total Members`,
    `Summit Members`,
    `Oxygen Used`,
    `Highest Point (m)`,
    `Weather Condition`,
    `Temperature (°C)`,
    `Wind Speed (km/h)`,
    `Snowfall (cm)`,
    `Rain (mm)`,
    `First Ascent Year`,
    `First Ascent Season`,
    `First Ascent Country`,
    `First Ascent Members`
FROM himalaya.gold.vw_master_expeditions;
```

##Peak Analysis

##Weather Patterns

```sql
CREATE OR REPLACE VIEW himalaya.gold.vw_weather AS
SELECT
    `Peak ID`,
    `Mountain`,
    `Height (m)`,
    `Region`,
    `Himalayan Range`,
    `Year`,
    `Season`,
    `Weather Date`,
    `Temperature (°C)`,
    `Rain (mm)`,
    `Snowfall (cm)`,
    `Snow Depth (m)`,
    `Wind Speed (km/h)`,
    `Weather Condition`,
    `Success`,
    `Deaths`,
    `Termination Reason`
FROM himalaya.gold.vw_master_expeditions;
```

##Death Analysis

```sql
CREATE OR REPLACE VIEW himalaya.gold.vw_deaths AS
SELECT
    `Expedition ID`,
    `Peak ID`,
    `Mountain`,
    `Height (m)`,
    `Region`,
    `Year`,
    `Season`,
    `Name`,
    `Nationality`,
    `Sex`,
    `Year of Birth`,
    `Hired`,
    `Sherpa`,
    `Death`,
    `Cause of Death`,
    `Death Date`,
    `Highest Point (m)`,
    `Temperature (°C)`,
    `Rain (mm)`,
    `Snowfall (cm)`,
    `Snow Depth (m)`,
    `Wind Speed (km/h)`,
    `Weather Condition`
FROM himalaya.gold.vw_master_members
WHERE `Death` = true;
```
